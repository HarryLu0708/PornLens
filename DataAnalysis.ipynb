## Extract Data from Detailed Infos

```python
import ast
```

```python
test_cell = unique_data["Detailed Infos"][1]
test_cell
```

```python
dict_test_cell = ast.literal_eval(test_cell)
dict_test_cell
```

```python
for key in dict_test_cell:
    print(key)
    print("value:",dict_test_cell[key])
```

```python
unique_data["Detailed Infos"] = unique_data["Detailed Infos"].fillna("missing")
```

```python
info_names = []
column = unique_data["Detailed Infos"]
for idx,d in enumerate(column):
    if d=="missing":
        print("missing")
        continue
    try:
        dict_test_cell = ast.literal_eval(d)
    except:
        print("error")
    for key in dict_test_cell:
        if key not in info_names:
            info_names.append(key)
info_names
```

```python
new_data = unique_data.reindex(columns=unique_data.columns.tolist()+info_names)
new_data.shape, new_data.head()
```

```python
info_names = []
processed_data = new_data
column = new_data["Detailed Infos"]

for idx,d in enumerate(column):
    try:
        dict_test_cell = ast.literal_eval(d)
    except:
        print("error")
    for key in dict_test_cell:
        processed_data.at[idx,key] = dict_test_cell[key]
processed_data.head()
```

```python
processed_data.isnull().mean()
```

## Extract Information from Social Medias

```python
social_medias = processed_data["Social Medias"]
dic = ast.literal_eval(social_medias[0])
dic
```

```python
sm_names = []
for sm in social_medias:
    try:
        dic = ast.literal_eval(sm)
    except:
        print("error")
    for key in dic:
        if key not in sm_names:
            sm_names.append(key)
sm_names
```

```python
general = ['Twitter', 'Snapchat', 'Instagram', 'Official Site','Amazon Wishlist']
social_media_columns = general + ["Other"]
social_media_columns
```

```python
processed_data = processed_data.reindex(columns=processed_data.columns.tolist()+social_media_columns)
processed_data.shape, processed_data.head()
```

```python
for idx,sm in enumerate(social_medias):
    try:
        dic = ast.literal_eval(sm)
    except:
        print("error")
    for key in dic:
        if key not in general:
            processed_data.at[idx,"Other"] = dic[key]
        else:
            processed_data.at[idx,key] = dic[key]
processed_data.head()
```

```python
processed_data.isnull().mean()
```

## Clean Last Month Rank

```python
def convert_to_data(s):
    new_s = s[1:-1]
    if new_s=="":
        return
    return int(new_s)
convert_to_data("")
```

```python
processed_data["Last Month Rank"] = processed_data["Last Month Rank"].apply(convert_to_data)
```
